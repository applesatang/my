# 작업 2유형

```python
import pandas as pd
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

print(display(x_train.head()))
# x_test.head()

# print(x_train.isnull().sum())
# print(x_train.info())
print(x_train.nunique())
#최빈값 = value_counts().index[0], 평균값 = mean()
e_m = x_train['Ever_Married'].value_counts().index[0]
g = x_train['Graduated'].value_counts().index[0]
p = x_train['Profession'].value_counts().index[0]
f = x_train['Family_Size'].value_counts().index[0]
w_e = x_train['Work_Experience'].mean()
v1 = x_train['Var_1'].value_counts().index[0]

x_train.Ever_Married = x_train['Ever_Married'].fillna(e_m)
x_train.Graduated = x_train['Graduated'].fillna(g)
x_train.Profession = x_train['Profession'].fillna(p)
x_train.Work_Experience = x_train['Work_Experience'].fillna(w_e)
x_train.Family_Size = x_train['Family_Size'].fillna(f)
x_train.Var_1 = x_train['Var_1'].fillna(v1)

x_test.Ever_Married = x_test['Ever_Married'].fillna(e_m)
x_test.Graduated = x_test['Graduated'].fillna(g)
x_test.Profession = x_test['Profession'].fillna(p)
x_test.Work_Experience = x_test['Work_Experience'].fillna(w_e)
x_test.Family_Size = x_test['Family_Size'].fillna(f)
x_test.Var_1 = x_test['Var_1'].fillna(v1)

# print(x_test.isnull().sum())
# print(x_train.isnull().sum())

# Family_Size 범주형 변경
x_train.Family_Size = x_train.Family_Size.astype('str')
x_test.Family_Size = x_test.Family_Size.astype('str')

train_drop = x_train.drop(columns = ['ID','Segmentation'])
test_drop = x_test.drop(columns = ['ID'])
y = x_train['Segmentation']

#더미화
train_dummies = pd.get_dummies(train_drop)
test_dummies = pd.get_dummies(test_drop)
#칼럼 순서 맞추기
test_dummies = test_dummies[train_dummies.columns]

from sklearn.model_selection import train_test_split
# help(train_test_split)
x_t, x_v, y_t, y_v = train_test_split(train_dummies,y,stratify = y, test_size = 0.33, random_state = 42)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 23)
rf.fit(x_t,y_t)

predict_train_label = rf.predict(x_t)
predict_val_label = rf.predict(x_v)

from sklearn.metrics import f1_score

print(f1_score(y_t,predict_train_label,average = 'macro'))
print(f1_score(y_v,predict_val_label,average = 'macro'))

pred_test = rf.predict(test_dummies)

result = pd.DataFrame({'ID' : x_test.ID , 'Segmentation' : pred_test })
result.to_csv('0030.csv',index = False)

check = pd.read_csv( '0030.csv')
print(display(check))
```

# 3유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p3_1_.csv')
print(display(df.head()))

result1 = round(df.groupby('학과').size().mean(),3)
print(result1)

# 카이제곱 

from scipy.stats import chi2_contingency

crosstab = pd.crosstab(df['학과'], df['성별'])
s, p , _, _ = chi2_contingency(crosstab)
result2 = round(s,3)
print(result2)

result3 = round(p,3)
print(result3)
print('귀무')
```

### 어느 학교에서 수학 시험을 본 학생 100명 중 60명이 60점 이상을 받았다. 이 학교의 수학 시험의 평균 점수가 50점 이상인지 95%의 신뢰수준에서 검정하려한다.

```python
import numpy as np
n = 100
p_hat = 0.6
p = 0.5
alpha = 0.05

#검정통계량 계산
z = round((p_hat -p) / np.sqrt(p*(1-p) /n),5)
print(z)
```

```python
from scipy.stats import norm
p_value = round(1-norm.cdf(z),3)
print(p_value)
print('대립')
```
