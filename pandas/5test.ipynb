# 제 1유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_1_.csv')
# print(display(df))
print(df.info())

df_filter = df.loc[(df['20L가격'] != 0) & (df['5L가격'] != 0)]
df_filter['차이가격'] = df_filter['20L가격'] - df_filter['5L가격']
result = df_filter.groupby(['시도명'])['차이가격'].mean().sort_values().to_frame()
print(result.info())
answer = round(result['차이가격'].max(),1)
print(result)
print(result.head(1))
# answer = result.iloc[:0,:1]
print(answer)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_2_.csv')
print(df.info())
df['height(m)'] = df['height(cm)']/100
df['BMI'] = df['weight(kg)'] / (df['height(m)'] *df['height(m)'])
print(df.info())

def bmi_check(x) :
    if x >= 25 :
        return '초고도 비만'
    elif x>= 23 :
        return '고도 비만'
    elif x>= 18.5 :
        return '정상'
    else : return '저체중'

df['비만여부'] = df.BMI.apply(bmi_check)

result1 = df.loc[(df['비만여부'] == '초고도 비만') | (df['비만여부'] == '저체중')].value_counts().sum()
print(result1)
```

# 제 2유형

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')
# display(train.head())
# display(test.head())

# print(train.info())
# print(train.isnull().sum())
# print(train.nunique())

train_drop = train.drop(columns = [('ID') , ('price')])
test_drop = test.drop(columns = ['ID'])
y = train['price']

tr_dm = pd.get_dummies(train_drop)
te_dm = pd.get_dummies(test_drop)
print(tr_dm.info())
print(te_dm.info())
te_dm = te_dm.reindex(columns = tr_dm.columns, fill_value =0 )
# import sklearn.ensemble
# print(dir(sklearn.ensemble))
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error 
# print(dir(sklearn.metrics))
rf = RandomForestRegressor(random_state = 23)
x_t, x_v, y_t, y_v = train_test_split(tr_dm, y, test_size = 0.33, random_state =42)
rf.fit(x_t,y_t)

pred = rf.predict(x_v)
print(mean_squared_error(y_v,pred, squared = False))

pred_test = rf.predict(te_dm)
result = pd.DataFrame({ 'ID' : test.ID , 'price' : pred_test})
print(result)
```
